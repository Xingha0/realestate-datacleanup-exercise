#### Ejercicio 01. ¿Cuál es la casa mas cara en el todo el dataset? (★☆☆)

#### Ejercicio 09. ¿ Son los precios promedios de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedio y escribe una conclusion acerca de ellos

```python
# 
valdemorillo_ds = ds[ds['level5'] == 'Valdemorillo']
galapagar_ds = ds[ds['level5'] == 'Galapagar']
valdemorillo_precio_med = valdemorillo_ds['price'].mean()
galapagar_precio_med = galapagar_ds['price'].mean()

print("Precio medio en Valdemorillo:", valdemorillo_precio_med, "€")
print("Precio medio en Galapagar:", galapagar_precio_med, "€")

if valdemorillo_precio_med == galapagar_precio_med:
    print("El precio medio de Valdemorillo y Galapagar son iguales.")
else:
    print("El precio medio de Valdemorillo y Galapagar son diferentes.")
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
fig, ax = plt.subplots()

precio_normal = []

for poblacion in cintsur_pop:
    csdat = cintsur[cintsur['level5'] == poblacion]
    precio = csdat['price']
    precio_norm = (precio - precio.min()) / (precio.max() - precio.min())
    precio_normal.append(precio_norm)

plt.hist(precio_normal, label = cintsur_pop)
plt.title("Histograma de precios normalizados del Cinturón Sur")
plt.xlabel("Precio Normalizado")
plt.ylabel("Frecuencia")
plt.legend(title="Población")
plt.show()
```
